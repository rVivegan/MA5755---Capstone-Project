```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assign categories with improved mapping
def assign_category(item_name):
    category_map = {
        # Beverages
        'Cappuccino': 'Beverages',
        'Cafe Latte': 'Beverages',
        'Espresso': 'Beverages',
        'Filter Coffee': 'Beverages',
        'Honey Cinnamon Coffee': 'Beverages',
        'Cafe Americano': 'Beverages',
        'Flat White': 'Beverages',
        'Cafe Mocha': 'Beverages',
        'Chai': 'Beverages',
        'White Mocha': 'Beverages',
        'Salted Caramel': 'Beverages',
        'Coconut Spiced Latte': 'Beverages',
        'Tea Pot': 'Beverages',
        'Slush': 'Beverages',
        'Frappe': 'Beverages',
        'Refresher': 'Beverages',
        'Iced Coffee': 'Beverages',
        'Mocha': 'Beverages',
        'Coffee': 'Beverages',
        
        # Bakery
        'Croissant': 'Bakery',
        'Garlic Bread': 'Bakery',
        'Sandwich': 'Bakery',
        'Toastizza': 'Bakery',
        'Toast': 'Bakery',
        'Tikka': 'Bakery',
        'Double Decker': 'Bakery',
        'Cheese Tomato': 'Bakery',
        'Corn N Cheese': 'Bakery',
        'Peri Peri': 'Bakery',
        
        # Desserts
        'Affogato': 'Desserts',
        'Brownie': 'Desserts',
        'Cheesecake': 'Desserts',
        'Fantasy': 'Desserts',
        'Sundae': 'Desserts',
        'Cake': 'Desserts',
        'Choco': 'Desserts',
        'Fudge': 'Desserts',
        'Nirvana': 'Desserts',
        'Sizzle': 'Desserts',
        'Red Velvet': 'Desserts',
        'Death By Chocolate': 'Desserts',
        'Truffle': 'Desserts',
        'Butterscotch': 'Desserts',
        'Gooey': 'Desserts'
    }
    
    for key in category_map:
        if key.lower() in item_name.lower():
            return category_map[key]
    return 'Other'

# Apply the category mapping to your dataset
data['Category'] = data['Item Name'].apply(assign_category)

# Calculate aggregates by category
category_stats = data.groupby('Category').agg({
    'Net Sales': 'sum',
    'Quantity': 'sum',
    'Discount': ['mean', 'sum'],
    'Rate': 'mean'
}).reset_index()

# Flatten the multi-index columns
category_stats.columns = [f'{col[0]}_{col[1]}' if col[1] != '' else col[0] for col in category_stats.columns]

# Create subplots for a more professional visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot Net Sales by Category
sns.barplot(x='Category', y='Net Sales_sum', data=category_stats, ax=ax1, palette='viridis')
ax1.set_title('Total Net Sales by Category')
ax1.set_xlabel('Category')
ax1.set_ylabel('Total Net Sales')
ax1.ticklabel_format(style='plain', axis='y')

# Plot Quantity by Category
sns.barplot(x='Category', y='Quantity_sum', data=category_stats, ax=ax2, palette='magma')
ax2.set_title('Total Quantity by Category')
ax2.set_xlabel('Category')
ax2.set_ylabel('Total Quantity')

plt.tight_layout()
plt.show()

# Display category stats for detailed analysis
print("Category Statistics Summary:")
print(category_stats[['Category', 'Net Sales_sum', 'Quantity_sum', 'Discount_mean', 'Rate_mean']])

# Calculate profitability metrics
category_stats['Avg_Revenue_Per_Item'] = category_stats['Net Sales_sum'] / category_stats['Quantity_sum']
category_stats['Discount_Impact'] = category_stats['Discount_sum'] / category_stats['Net Sales_sum'] * 100

# Display profitability metrics
print("\nProfitability Metrics by Category:")
print(category_stats[['Category', 'Avg_Revenue_Per_Item', 'Discount_Impact']])

# Optional: Plot Top 10 items by sales
top_items = data.groupby('Item Name')['Net Sales'].sum().nlargest(10).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Net Sales', y='Item Name', data=top_items, palette='Blues_d')
plt.title('Top 10 Items by Net Sales')
plt.xlabel('Net Sales')
plt.ylabel('Item Name')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load cleaned dataset/content/CCD_Dataset.csv
#df = pd.read_csv('/content/ccd_data.csv')

# Calculate total sales per item
item_sales = df.groupby('Item Name')['Net Sales'].sum().sort_values(ascending=False).reset_index()
item_sales['Cumulative Sales'] = item_sales['Net Sales'].cumsum()
item_sales['Cumulative Percentage'] = item_sales['Cumulative Sales'] / item_sales['Net Sales'].sum() * 100

# Assign ABC categories
item_sales['Category'] = 'C'
item_sales.loc[item_sales['Cumulative Percentage'] <= 80, 'Category'] = 'A'
item_sales.loc[(item_sales['Cumulative Percentage'] > 80) & (item_sales['Cumulative Percentage'] <= 95), 'Category'] = 'B'

item_sales[['Item Name', 'Net Sales', 'Category']].to_csv('abc_analysis.csv', index=False)
print("ABC analysis saved as 'abc_analysis.csv'")
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(30, 10))
sns.barplot(data=item_sales, x='Item Name', y='Net Sales', hue='Category', dodge=False, palette='Set2')
plt.title('ABC Analysis: Net Sales by Item Category')
plt.xlabel('Item Name')
plt.ylabel('Net Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
# plt.figure(figsize=(30, 10))
# plt.plot(item_sales['Item Name'], item_sales['Cumulative Percentage'], marker='o', linestyle='-')
# plt.axhline(80, color='green', linestyle='--', label='A/B Threshold (80%)')
# plt.axhline(95, color='orange', linestyle='--', label='B/C Threshold (95%)')
# plt.title('ABC Analysis: Cumulative Sales Percentage')
# plt.xlabel('Item Name')
# plt.ylabel('Cumulative Sales (%)')
# plt.xticks(rotation=45)
# plt.legend()
# plt.tight_layout()
# plt.show()
category_counts = item_sales['Category'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', colors=['green', 'orange', 'red'])
plt.title('Item Distribution by ABC Category')
plt.show()
```
